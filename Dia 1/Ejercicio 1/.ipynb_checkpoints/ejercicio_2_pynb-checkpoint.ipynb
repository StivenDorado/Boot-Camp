# "Registro de Ventas Mensuales de Electrónica (Enero - Abril 2024)"




Este dataset contiene información detallada sobre las ventas de productos electrónicos durante los primeros cuatro meses del año 2024. Los datos incluyen la fecha de la venta, el ID de la transacción, el producto vendido (Smartphone, Tablet, Laptop, Headphones), la cantidad vendida, el precio unitario de cada artículo y el total de la venta. Los productos más vendidos son los smartphones, seguidos por tablets, laptops y headphones, reflejando las fluctuaciones en las cantidades y precios a lo largo del período. Además, algunos registros presentan valores faltantes o negativos en la cantidad o total de ventas, lo que puede indicar devoluciones o ajustes.

Este dataset puede ser útil para análisis de tendencias de ventas, comportamiento de precios y productos más demandados, así como para la gestión de inventarios y estrategia de ventas.

Realiza una inspección inicial para comprender el estado de los datos:

Paso 4: Limpieza de Datos
Eliminar Duplicados: Si hay filas duplicadas, eliminarlas:

```python
df = df.drop_duplicates()
df
```

Tratar con valores nulos: Dependiendo de las columnas, puedes hacer lo siguiente:
Para Cantidad y Precio_Unitario, puedes reemplazar los valores nulos con la media o la mediana de la columna.
Para Producto o ID_Venta, si hay valores faltantes, puedes optar por eliminarlos o completarlos con algún valor estándar (como "Desconocido")

```python
# Reemplazar valores nulos en 'Cantidad' y 'Precio_Unitario' con la mediana
df['Cantidad'] = df['Cantidad'].fillna(df['Cantidad'].median())
df['Precio_Unitario'] = df['Precio_Unitario'].fillna(df['Precio_Unitario'].median())

# Reemplazar valores nulos en 'Producto' con "Desconocido"
df['Producto'] = df['Producto'].fillna('Desconocido')

```

```python
print(df.isnull().sum())
```

Corregir formatos de fecha: Asegúrate de que las fechas estén en un formato adecuado. Si hay registros con formatos inconsistentes, conviértelos:

```python
df['Fecha'] = pd.to_datetime(df['Fecha'], errors='coerce')  # Los valores incorrectos se convertirán en NaT
```

Eliminar o corregir valores incorrectos en Cantidad y Precio_Unitario: Si encuentras valores negativos o fuera de lugar, puedes corregirlos de la siguiente manera:

```python
df = df[df['Cantidad'] > 0]  # Eliminar registros con cantidades negativas o cero
df = df[df['Precio_Unitario'] > 0]  # Eliminar registros con precios negativos o cero
```

Recalcular el Total: Si el campo Total tiene valores incorrectos, recalcularlo multiplicando Cantidad por Precio_Unitario:

```python
df['Total'] = df['Cantidad'] * df['Precio_Unitario']
```

Paso 5: Verificación y Resultados
Verificar los resultados:

```python
print(df.head())  # Verifica cómo quedaron las primeras filas después de la limpieza
print(df.describe())  # Revisa las estadísticas descriptivas después de la limpieza
```

Visualización: Puedes generar algunas gráficas para explorar los datos limpios. Por ejemplo, para analizar la distribución de los totales de ventas:

```python
# Agrupar los datos por Producto y calcular el total de ventas para cada uno
ventas_por_producto = df.groupby('Producto')['Total'].sum().sort_values(ascending=False)

# Crear la gráfica de barras
plt.figure(figsize=(10, 6))  # Ajustar el tamaño de la figura
ventas_por_producto.plot(kind='bar', color='skyblue')
plt.title('Totales de Ventas por Producto')
plt.xlabel('Producto')
plt.ylabel('Total de Ventas ($)')
plt.xticks(rotation=45, ha='right')  # Rotar las etiquetas del eje X para mejor lectura
plt.tight_layout()  # Ajustar el layout para que no se corten las etiquetas
plt.show()

```

Explicación de la Gráfica:
Eje X (horizontal): Representa los diferentes productos que se han vendido.
Eje Y (vertical): Muestra el total de ventas en dólares o la moneda que corresponda.
Barra: Cada barra representa el total de ventas para un producto. Los productos con más ventas tendrán barras más largas, y los productos con menos ventas tendrán barras más cortas.
Rotación de etiquetas en el eje X: Las etiquetas de los productos pueden ser largas, por lo que las rotamos 45 grados para que sea más fácil leerlas.
Resultado Esperado:
La gráfica te permitirá ver rápidamente qué productos están generando más ingresos. Esto puede ser útil para tomar decisiones sobre inventarios, promociones o estrategias de ventas.
